```python
#importing dependencies
import pandas as pd
import numpy as np
import requests
import json
import matplotlib.pyplot as plt
import random
import time
from scipy.stats import linregress
```

```python
#dependencies and key
from api_key import api_key
from citipy import citipy
```

```python
#saving info for api calls
base_url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

#query_url = "api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units={units}"

query_url = f"{base_url}&units={units}&appid={api_key}&q="
print(query_url)
```

```python
#this is for the range of lon and lat
latitude_range = (-90, 90)
longitude_range = (-180, -80)

#setting up the list and the random lat and lon
lat_and_lon = []
cities = []

latitude1 = np.random.uniform(low = -90, high = 90, size = 1500)
longitude1 = np.random.uniform(low = -180, high = 180, size = 1500)

lat_and_lon = zip(latitude1, longitude1)

#using citipy to get cities
for lat1_lon1 in lat_and_lon:
    city = citipy.nearest_city(lat1_lon1[0], lat1_lon1[1]).city_name
    
    if city not in cities:
        cities.append(city)
        
print(len(cities))

```

```python
#creating a list for the data
lat_data = []
lng_data = []
temp_data = []
cloud_data = []
wind_data = []
humidity = []
city_name = []
country = []

#Created a loop that will go through the cities that generated
#missing results message comes up as if there is an error... i guess there is
#no data comes out of what i did

counter = 1

for city in cities:
    
    try:
        print(query_url + city)
        response = requests.get(query_url + city).json()
        lat_data.append(response['coord']['lat'])
        lng_data.append(response['coord']['lon'])
        temp_data.append(response['main']['temp'])
        cloud_data.append(response['clouds']['all'])
        wind_data.append(response['wind']['speed'])
        humidity.append(response['main']['humidity'])
        city_name.append(response['name'])
        country.append(response['sys']['country'])
        current_city = response['name']
        print(f"Retrieving results for {counter} in: {current_city}")
        
        counter = counter + 1
        
        


    except(KeyError): 
        print("Missing results, no city data: Skipped")
        print("-----------")
        
    time.sleep(1)
        
    
    
```

```python
#putting the info i pulled into a DFs
cities_info = pd.DataFrame({"City": city_name,
                           "Country": country,
                           "Temp": temp_data,
                           "Humidity": humidity,
                           "Cloudiness": cloud_data,
                           "Wind Speed": wind_data,
                           "Lat": lat_data,
                           "Lng": lng_data})
cities_info.head()
```

```python
cities_info.to_csv('OutputData/city_weather.csv', index=False, header=True)
```

```python
#scatter plot for temp vs lat
plt.scatter(cities_info["Temp"], cities_info["Lat"], marker="o", facecolors="red", edgecolors="black")
plt.title("Temperature vs Latitude")
plt.xlabel("Temperature")
plt.ylabel("Latitude")

x_values = cities_info["Temp"]
y_values = cities_info["Lat"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.annotate(line_eq,(6,10),fontsize=15, color="black")
plt.plot(x_values, regress_values, "b-")

plt.savefig('OutputData/temp_vs_lat.png')
```
